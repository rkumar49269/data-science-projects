Importing the dependencies.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

This is the case of under sampling, so we are building a sample dataset containing similar distribution of normal transactions and fraudulent transactions.

```python
legit_samples = legit.sample(n = 492)
```

```python
# Concatenation two dataframes
new_dataset = pd.concat([legit_samples, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset.describe()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.describe()
```

```python
# Separating X and Y from dataset
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
#Splitting the dataset into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 2, stratify=Y)
```

```python
# loading the model
mdl = LogisticRegression()
```

```python
# training model.
mdl.fit(X_train, Y_train)
```

```python
#Evaluation
#Accuracy for training data
train_prediction = mdl.predict(X_train)
train_accuracy = accuracy_score(train_prediction, Y_train)
print("Accuracy for train: ", train_accuracy)
```

```python
#Accuracy for testing data
test_prediction = mdl.predict(X_test)
test_accuracy = accuracy_score(test_prediction, Y_test)
print("Accuracy for train: ", test_accuracy)
```
