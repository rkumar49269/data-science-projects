```python
#Data Preprocessing
sales_data.head()
```

```python
sales_data['Item_Fat_Content'].value_counts()
```

```python
sales_data.replace({'Item_Fat_Content': {'low fat': 'Low Fat', 'LF': 'Low Fat', 'reg': 'Regular'}}, inplace=True)
```

```python
sales_data['Item_Fat_Content'].value_counts()
```

```python
# Lable Encoding
encode = LabelEncoder()
```

```python
sales_data['Item_Identifier'] = encode.fit_transform(sales_data['Item_Identifier'])
sales_data['Item_Fat_Content'] = encode.fit_transform(sales_data['Item_Fat_Content'])
sales_data['Item_Type'] = encode.fit_transform(sales_data['Item_Type'])
sales_data['Outlet_Identifier'] = encode.fit_transform(sales_data['Outlet_Identifier'])
sales_data['Outlet_Size'] = encode.fit_transform(sales_data['Outlet_Size'])
sales_data['Outlet_Location_Type'] = encode.fit_transform(sales_data['Outlet_Location_Type'])
sales_data['Outlet_Type'] = encode.fit_transform(sales_data['Outlet_Type'])
```

```python
sales_data.head()
```

```python
#Splitting data into features and target
X = sales_data.drop(columns = 'Item_Outlet_Sales', axis = 1)
Y = sales_data['Item_Outlet_Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model training
mdl = XGBRegressor()
```

```python
mdl.fit(X_train, Y_train)
```

```python
# Prediction and error score on training data
train_prediction = mdl.predict(X_train)
train_r2score = metrics.r2_score(Y_train, train_prediction)
print("Error by r2 score: ", train_r2score)
```

```python
# Prediction and error score on testing data
test_prediction = mdl.predict(X_test)
test_r2score = metrics.r2_score(Y_test, test_prediction)
print("Error by r2 score: ", test_r2score)
```
